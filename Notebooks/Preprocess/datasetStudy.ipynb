# Check for duplicated captions

```python
# group by 'body' column and count occurrences
counts_caption = df.groupby('caption').size().reset_index(name='count')

# select only rows where count is greater than 1
duplicates_caption = counts_caption.loc[counts_caption['count'] > 1]

print(duplicates_caption['count'].sum())
duplicates_caption.sort_values('count', ascending=False)
```

# Check for duplicated body texts

```python
# group by 'body' column and count occurrences
counts = df.groupby('body').size().reset_index(name='count')

# select only rows where count is greater than 1
duplicates = counts.loc[counts['count'] > 1]

print(duplicates['count'].sum())
duplicates.sort_values('count', ascending=False)
```

# Study of the length of texts

```python
def obtain_len(text):
    words=text.split()
    return len(words)
df2=df.copy()
df2['len_headline']= df2['headline'].apply(obtain_len)
df2['len_abstract']= df2['abstract'].apply(obtain_len)
df2['len_caption']= df2['caption'].apply(obtain_len)
df2['len_body']= df2['body'].apply(obtain_len)
```

```python
df2.sort_values('len_headline', ascending=False)
```

```python
df2.sort_values('len_abstract', ascending=False)
```

```python
df2.sort_values('len_caption', ascending=False)
```

```python
df2.sort_values('len_body', ascending=False)
```
